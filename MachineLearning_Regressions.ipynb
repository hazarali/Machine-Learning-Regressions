```python
x=data.horsepower.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
yhead=lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,yhead)
```
